# __Análisis bi-variado__

Un análisis bi-variado exploratorio se presenta cuando se tienen registros de al menos dos características acerca del mismo fenómeno o que definen una entidad y que guardan una relación.

- __Relación estadística__
Dos variables x e y están relacionadas estadísticamente cuando conocida la primera se puede estimar aproximadamente el valor de la segunda.

- __Relación funcional__
Dos variables x e y están relacionadas funcionalmente cuando conocida la primera se puede saber con exactitud el valor de la segunda.

### __Variable estadística bidimensional__
Una variable bidimensional es una variable en la que cada individuo está definido por un par de caracteres, $(X, Y)$.

Estos dos caracteres son a su vez variables estadísticas en las que sí existe relación entre ellas, una de las dos variables es la variable independiente y la otra variable dependiente.

### __Distribuciones bidimensionales__
Son aquellas en las que a cada individuo le corresponden los valores de dos variables, las representamos por el par $(x_i, y_i)$.

Si representamos cada par de valores como las coordenadas de un punto, el conjunto de todos ellos se llama nube de puntos o diagrama de dispersión.

Sobre la nube de puntos puede trazarse una recta que se ajuste a ellos lo mejor posible, llamada recta de regresión.

## __Covarianza__

### __Definición__
 
La __covarianza__ de una muestra bidimensional es la media aritmética de los productos de las desviaciones de cada una de las variables respecto a sus medias respectivas. Es decir

$$s_{xy} = \frac{1}{n} \sum_{i=1}^n (x_i-\bar x)(y_i-\bar y)$$

La _covarianza_ se suele representar por $s_{xy}$. En algunas ocasiones también se suele representar mediante $\sigma_{xy}$, aunque no es recomendable.

Otra manera equivalente de calcular la covarianza es 
 

$$ s_{xy} = \frac{1}{n} \sum_{i = 1}^{n}{x_i y_i} - \overline{x} \cdot \overline{y}$$

El resultado debe ser el mismo sin importar la fórmula utilizada 
- observa que el producto $\overline{x} \cdot \overline{y}$ no se divide por $n$.

---

### __Interpretación y propiedades__
 
- La covarianza indica el sentido de la correlación entre las variables

- Si $s_{xy}> 0$ la correlación es directa.

- Si $s_{xy}< 0$ la correlación es inversa.


__Nota:__ La covarianza presenta como inconveniente, el hecho de que su valor depende de la escala elegida para los ejes. Es decir, la covarianza variará si expresamos la altura en metros o en centímetros. También variará si el dinero lo expresamos en euros o en dólares.

__Algortimo covarianza__

- Se define la función que recibe dos vectores __X__, __Y__
- Observando la fórmula de covarianza, se debe calcular __previamente__ el valor del promedio de $x$: como $\bar x$ y luego el promedio de $y$: como $\bar y$
- Se realiza un recorrido en donde se __acumula__ la multiplicación cada valor $x_i, y_i$ menos su respectiva media $(\bar x, \bar y)$
- Se calcula la covarianza $s_{xy}$ y se devuelve el respectivo valor




````
# haga la función del promedio y desviación de una variable aleatoria

def promedio(materia):

  #Sumatoria de todos los valores
  valores = sum(materia)
  #Cantidad de datos
  n_datos = len(materia)
  
  return valores/n_datos

def desviacion(materia):

  numerador = 0
  n_datos = len(materia)

  #Sumatoria nota menos la media elevado todo al cuadrado
  for nota in materia:
    numerador = numerador + pow((nota - promedio(materia)),2)

  #Se aplica la raiz de la sumatoria y se divide entre en numero de datos
  return sqrt(numerador/n_datos)
````

## __Coeficiente de correlación lineal__

El coeficiente de correlación lineal es el cociente entre la covarianza y el producto de las desviaciones típicas de ambas variables. El coeficiente de correlación lineal se representa mediante la letra $\rho$.
$$\rho_{xy} = \frac{\sigma_{xy}}{\sigma_x \sigma_y} = \frac{\sum_{i = 1}^{n}{x_i y_i} - n\overline{x} \overline{y}}{(n-1) s_x s_y}$$
<br>
en donde, $\sigma_{xy}$ es la covarianza, $s_x$ es la desviación estándar de $x$ y $s_y$ es la desviación estándar de $y$ 
- En la siguiente figura se muestran ejemplos de gráficas de dispersión entre dos variables y su correspondiente valor de correlación lineal. Fuente [wikipedia](https://es.wikipedia.org/wiki/Coeficiente_de_correlaci%C3%B3n_de_Pearson)

![ejemplos de correlación](https://upload.wikimedia.org/wikipedia/commons/thumb/d/d4/Correlation_examples2.svg/1920px-Correlation_examples2.svg.png)

__Propiedades__
 

1. El coeficiente de correlación no varía al hacerlo la escala de medición. Es decir, si expresamos la altura en metros o en centímetros el coeficiente de correlación no varía.

2. El signo del coeficiente de correlación es el mismo que el de la covarianza.
    - Si la covarianza es positiva, la correlación es directa.

    - Si la covarianza es negativa, la correlación es inversa.

    - Si la covarianza es nula, no existe correlación.

3. El coeficiente de correlación lineal es un número real comprendido entre $-1\leq r\leq 1$

4. Si el coeficiente de correlación lineal toma valores cercanos a -1 la correlación es fuerte e inversa, y será tanto más fuerte cuanto más se aproxime a -1.

5. Si el coeficiente de correlación lineal toma valores cercanos a 1 la correlación es fuerte y directa, y será tanto más fuerte cuanto más se aproxime a 1.

6. Si el coeficiente de correlación lineal toma valores cercanos a 0, la correlación es débil.

7. Si $r=1$ ó $r=-1$, los puntos de la nube están sobre la recta creciente o decreciente. Entre ambas variables hay dependencia funcional.



## __Ejemplo:__

- Usando los datos de 'mpg' de Seaborn calcule el coeficiente de correlación entre las variables: 'mpg', 'displacement', 'horsepower' y 'weight'. Genere una matriz de correlación de 4x4 con las variables mencionadas.
- Muestre las gráficas de dispersión de estas variables y ponga en el título el coeficiente de correlación.

>__Algoritmo__
- Calcular la covarianza de X, Y, $\sigma_{xy}$
- Calcular la desviación estándar $X$, $\sigma_x$
- Calcular la desviación estándar $Y$, $\sigma_y$
- Hallar y devolver el valor del coeficiente de correlación
$$\rho_{xy} = \frac{\sigma_{xy}}{\sigma_x \sigma_y} = \frac{\sum_{i = 1}^{n}{x_i y_i} - n\overline{x} \overline{y}}{(n-1) s_x s_y}$$


````
# programe la función que calcula el coeficiente de correlación y pruebe

def coeficiente_correlacion(x,y):

  #Covarianza
  cov_xy = covarianza(x,y)

  #desviacion de x
  desv_x = desviacion(x)
  
  #desviacion de y
  desv_y = desviacion(y)

  #Coeficiente de correlación
  p_xy = cov_xy / (desv_x * desv_y)

  return p_xy
````

# __Regresión Lineal__

El objetivo de la regresión lineal es hallar los parámetros de la ecuación de línea recta que mejor se ajustan a una nube de puntos dada por las observaciones registradas de dos variables. La noción del mejor ajuste de una línea respecto de sus obervaciones es algo en lo que no se tiene consenso, por esta razón existen en la actualidad varios modelos de regresión lineal.  En general se busca que la distancia entre los puntos de la recta y sus observaciones sea mínimo, sin embargo en la distribución de estas distancias se tiene un sesgo y una varianza. Existe el modelos no sesgado, es decir que la recta minimiza la distancia de la obervación respecto a su valor en $y$ y también se tienen modelos que buscan reducir la varianza.

La recta de regresión, de bajo sesgo, que pasa por el punto ${(\overline{x}, \overline{y})}$ o centro de gravedad conformado por las medias aritméticas de los valores de $x$ y $y$ puede encontrase de la siguiente forma:

> __Recta de regresión de Y sobre X:__
- La recta de regresión de $y$ sobre $x$ se utiliza para estimar los valores de la $y$ a partir de los de $x$.

$$y-\overline{y}=\displaystyle\frac{s_{xy}}{s^{2}_{x}}(x-\overline{x})$$

> - La pendiente de la recta es el cociente entre la covarianza $s_{xy}$ y la varianza $s^{2}_{x}$ de la variable $x$.

> __Recta de regresión de X sobre Y__
- La recta de regresión de $x$ sobre $y$ se utiliza para estimar los valores de $x$ a partir de los de $y$.

$$x-\overline{x}=\displaystyle\frac{s_{xy}}{s^{2}_{y}}(y-\overline{y})$$

> - La pendiente de la recta es el cociente entre la covarianza $s_{xy}$ y la varianza $s^{2}_{y}$ de la variable $y$.
- Si la correlación es nula, esto es $r=0$, las rectas de regresión son perpendiculares entre si.


 

## __Ejemplo de recta de regresión:__

Con los datos de calificaciones entre matemáticas y física de un ejemplo anterior:
- Calcule los parámetros de las líneas de regresión de X sobre Y y de Y sobre X.
- Genera la gráfica de dispersión de las dos variables
- Pinte las dos rectas sobre la dispersión de los datos

>__Algoritmo solución__
1. Cálculo de parámetros de la recta de regresión de Y sobre X
    - Implementar la ecuación $y-\overline{y}=\displaystyle\frac{s_{xy}}{s^{2}_{x}}(x-\overline{x})$
    - Tomar la ecuación y convertirla a la forma $y = m\cdot x+b$ y obtener lapendiente $m$ y el punto de corte $b$ 
    $$m = \frac{s_{xy}}{s_x^2}$$ 
    $$b = \bar y - \frac{s_{xy}}{s_x^2}\bar x$$
2. Cálculo de parámetros de la recta de regresión de X sobre Y
    - Implementar la ecuación $x-\overline{x}=\displaystyle\frac{s_{xy}}{s^{2}_{y}}(y-\overline{y})$
    - Tomar la ecuación y convertirla a la forma $y = m\cdot x+b$ y obtener la 
    $$m = \frac{s_y^2}{s_{xy}}$$ 
    $$b = \bar y - \frac{s_y^2}{s_{xy}}\bar x$$
3. Generar la gráfica de dispersión entre X, Y
4. Pintar las rectas dentro de la gráfica de dispersión
    - Obtener la imagen $y$ del máximo y el mínimo de $x$
    - Pintar línea desde el mínimo y hasta el máximo


````
# Programe la función que calcula los parámetros de las rectas de regresión y grafique

def rectas_de_regresion(x, y, texto, texto_x = 'Materia 1', texto_y = 'Materia 2'):

  #Pendiente Y sobre X
  mx = (covarianza(x, y)) / (pow(desviacion(x), 2))

  #Pendiente X sobre Y
  my = (pow(desviacion(y), 2)) / (covarianza(x, y))

  #Arrays tipo numpy para operaciones matematicas
  x_r = promedio(x)
  y_r = promedio(y)

  #Regresion Y sobre X
  bx = y_r - (mx * x_r)

  #Regresion X sobre Y
  by = y_r - (my * x_r)

  #Ecuación de la recta de regresion Y sobre X
  yx = mx * np.array(x) + bx

  #Ecuación de la recta de regresion X sobre Y
  yy = my * np.array(x) + by

  # Crea la figura y los objetos axes
  fig, ax = plt.subplots(1, figsize=(9, 6))

  # Crea el grafico de disperción de la física
  ax.scatter(x, y,
            color='blue',   # Color of the dots
            s=80,          # Size of the dots
            alpha=0.75,     # Alpha/transparency of the dots (1 is opaque, 0 is transparent)
            linewidths=2)    # Size of edge around the dots

  #Regresion Y sobre X
  plt.plot(x, yx,
            color='mediumblue'   ,    # Color of the line
            alpha=0.75,         # Alpha/transparency of the dots (1 is opaque, 0 is transparent)
            linewidth=4)        #Line widht
           
  #Regresion X sobre Y
  plt.plot(x, yy,
            color='indigo',    # Color of the line
            alpha=0.75,          # Alpha/transparency of the dots (1 is opaque, 0 is transparent)
            linewidth=4)         #Line widht

  #Gráfica los títulos de las líneas y luego los puntos en el gráfico de dispersión
  ax.legend([('Regresion ' + texto_y + ' sobre ' + texto_x), 
             ('Regresion ' + texto_x + ' sobre ' + texto_y), 
             ('Dispersión(' + texto_x + ', ' + texto_y + ')')])
  
  #Texto del eje x
  plt.xlabel(texto_x)
  #Texto del eje y
  plt.ylabel(texto_y)

  #Título de la gráfica
  plt.title('Diagrama de dispersión de ' + texto)

  # Show
  plt.show()
````
