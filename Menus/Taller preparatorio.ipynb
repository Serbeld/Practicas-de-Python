```python
"""
El algoritmo debe iniciar con un menú como el que se muestra a continuación:
1. Ingresar un paciente.
2. Buscar un paciente.
3. Informe de pacientes.
4. Salir.
"""

from Modulo import buscar_un_paciente

condicion = 0
diccionario_general = {} 
paciente_contributivo = 0
paciente_subsidiado = 0
contador_de_hombres = 0
contador_de_mujeres = 0

while(condicion != '4'):
    
    print("""Menu
1. Ingresar un paciente.
2. Buscar un paciente.
3. Informe de pacientes.
4. Salir.
    """)
    
    condicion = input("Seleccione una opcion en el menu:")
    
    if(condicion == '1'):
        #Al seleccionar la opción 1. Ingresar un paciente, se debe preguntar si el paciente pertenece al
        #régimen contributivo (EPS) o subsidiado (SISBEN) (usar un menú para esta opción).
        
        #Respuesta del usuario
        regimen = input("""
A que regimen pertenece el paciente:
A. Régimen contributivo (EPS)
B. Subsidiado (SISBEN)
""")
        print()
        
        while(regimen != 'A' and regimen != 'B'):
            
            print()
            print("Ingrese un valor entre A y B")
            print()
            
            #Respuesta del usuario
            regimen = input("""
A que regimen pertenece el paciente:
A. Régimen contributivo (EPS)
B. Subsidiado (SISBEN)
""")
        
        if (regimen == 'A'):
            EPS = input("A que EPS pertenece el paciente: ")
            print()
            print("El paciente de regimen contributivo pertenece a la EPS " + str(EPS))
            paciente_contributivo = paciente_contributivo + 1
            #El numero del paciente se le asigna el numero del paciente contributivo si el valor ingresado por el paciente es contributivo
            numero_del_paciente = paciente_contributivo
            
        
        elif (regimen == 'B'):
            print()
            print("El paciente de regimen subsidiado pertenece")
            
            EPS = 'SISBEN'
            
            paciente_subsidiado = paciente_subsidiado + 1
            
            #El numero del paciente se le asigna el numero del paciente subsidiado si el valor ingresado por el paciente es subsidiado
            numero_del_paciente = paciente_subsidiado
            
        print()
        
        """
Se deben pedir los siguientes datos del paciente: Nombre, Documento de identidad (Id),
Género, Perfil Lipídico (ingresar los valores de Colesterol Total - COLT, HDL, LDL y
Triglicéridos - TRIG) y Diabetes (ingresar los valores de HbA1c y Glucosa) y almacenarlos en
diccionarios anidados como se muestra en el ejemplo. (ver imagen 1)"""
        
        Nombre = input("Ingrese el nombre del paciente: ")
        Id = int(input("Ingrese el Documento de identidad del paciente: "))
        Genero = input("Ingrese el Género (hombre o mujer) del paciente: ")
        COLT = int(input("Ingrese el Colesterol del paciente: "))
        HDL = int(input("Ingrese el HDL del paciente: "))
        LDL = int(input("Ingrese el LDL del paciente: "))
        TRIG = int(input("Ingrese los Triglicéridos del paciente: "))
        diabetes = float(input("Ingrese los valores de HbA1c del paciente: "))
        Glucosa = int(input("Ingrese la Glucosa del paciente: "))
        
        if (Genero == 'hombre'):
            contador_de_hombres = contador_de_hombres + 1
            
        if (Genero == 'mujer'):
            contador_de_mujeres = contador_de_mujeres + 1
        
        #Corresponde a un paciente ingresado
        diccionario_datos = {}
        diccionario_datos['Nombre'] = Nombre
        diccionario_datos['Id'] = Id
        diccionario_datos['EPS'] = EPS
        diccionario_datos['Sexo'] = Genero
        diccionario_datos['Perfil lipídico'] = {'COLT':COLT, 'HDL':HDL, 'LDL':LDL, 'TRIG':TRIG}
        diccionario_datos['Diabetes'] = {'HbA1c':diabetes,'Glucosa':Glucosa}
        
        #print(diccionario_datos)
        
        #Todos los datos de los pacientes
        diccionario_general[('EPS-' + str(EPS) +('-') + str(numero_del_paciente))] = diccionario_datos
        
        #print(diccionario_general)
        
        
    if(condicion == '2'):
        
        codigo_a_buscar = input('Ingrese el código del paciente: ')
        buscar_un_paciente(codigo_a_buscar, diccionario_general)
        print()
        
    if(condicion == '3'):
        
        print()
        print("***** Información de regimen *****")
        print("Contributivo: " + str(paciente_contributivo))
        print("Subsidiado: " + str(paciente_subsidiado))
        print("***** Información de regimen *****")
        print("Cantidad de hombres: " + str(contador_de_hombres))
        print("Cantidad de mujeres: " + str(contador_de_mujeres))
        print()
              
    if(condicion != '1' and condicion != '2' and condicion != '3' and condicion != '4'):
        
        print()
        print("Alerta: El valor ingresado no corresponte a ninguna de las opciones del menu")
        print()
```
