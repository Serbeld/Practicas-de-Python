# **¿Cuales estilos de cerveza representa la cerveza más amarga y menos amarga según su grado de IBU?**

Para resolver el problema se propone un ordenamiento de los datos referentes al archivo beers.csv con respecto a su nivel de IBU y denotar el estilo de cerveza que se encuentre al principio y al final de la lista.

```
# Se importan las librerias
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Se lee el archivo CSV con la libreria pandas
data = pd.read_csv('/content/drive/My Drive/Codigo_Mercedes/beers.csv')

#imprime el nombre de las columnas
names = data.columns[:]
print(names)
```

```
#Agrupar por cantidad de IBU y estilo de cerveza
IBU = data.groupby(['ibu','style'])
IBU['ibu','style'].describe()
```

```
fake = pd.DataFrame({'IBU' : data['ibu'].values, 'Estilos de cerveza': data['style'].values})

#Dimensiones a la figura
plt.figure(figsize=(10, 25))

fig = sns.barplot(x = 'IBU', y = 'Estilos de cerveza', data = fake)
```

```
#Menos amarga
Menos_amarga = pd.DataFrame(IBU)[0][0]

print(Menos_amarga)
```

```
#Mas amarga
Mas_amarga = pd.DataFrame(IBU)[0][IBU.ngroups-1]
print(Mas_amarga)
```
